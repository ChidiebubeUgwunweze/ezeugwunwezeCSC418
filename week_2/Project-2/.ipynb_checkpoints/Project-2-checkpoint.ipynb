```python
# Instructions for the application

#Importing necessary libraries
import cv2
from matplotlib import pyplot as plt
import numpy as np
import os

# Initializing directory containing the images
directory = r'C:\Users\NEW\Documents\SCHOOL\Year_4_semester_2\CSC 418\Practical\week_2\Project-2\img'

# Change working directory to directory containing image of all classmates
os.chdir(directory)


#Function definitions
def userinput(username, matno):
    return username, matno

#Definintion of Enhancement functions
def visualisecolorspaces(image, username):
    B, G, R = cv2.split(image)
    
    #Plot Original image
    plt.subplot(2, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    #Plot Blue image
    plt.subplot(2, 2, 2)
    plt.title("Blue")
    plt.imshow(B)
    
    #Plot Green image
    plt.subplot(2, 2, 3)
    plt.title("Green")
    plt.imshow(G)
    
    #Plot Red image
    plt.subplot(2, 2, 4)
    plt.title("Red")
    plt.imshow(R)
    
def additionofimage(image1, username, image2 = cv2.imread('megamind.jpeg')):
    image1 = cv2.resize(image1, (500,400))
    image2 = cv2.resize(image2, (500, 400))
    
    # cv2.addWeighted with applied parameters
    addImage = cv2.addWeighted(image1, 0.5, image2, 0.6, 0)
    
    #Plot the first image
    plt.subplot(1, 3, 1)
    plt.title(username)
    plt.imshow(image1)
    
    #Plot second image
    plt.subplot(1, 3, 2)
    plt.title("megamind")
    plt.imshow(image2)
    
    #Plot added image
    plt.subplot(1, 3, 3)
    plt.title(username + " and megamind")
    plt.imshow(addImage)
    
    
def brightness_contrast(image, username):
    #Plot the original image
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    #Adjust the brightness and contrast
    brightness = 5
    contrast = 1.5
    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)
    
    #Save the image
    #cv2.imwrite('bright_contrast_' + username +'.jpg', image2)
    
    #Plot the contrast image
    plt.subplot(1, 2, 2)
    plt.title("Brightness & contrast")
    plt.imshow(image2)
    plt.show()
    
def sharpening(image, username):
    #Plot the original image
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    #Create the sharpening kernel
    kernel = np.array([[0, -1, 0],[-1, 5, -1],[0, -1, 0]])
    
    # Sharpen the image
    sharpened_image = cv2.filter2D(image, -1, kernel)
    
    #Save the image
    #cv2.imwrite('sharpened_' + username +'.jpg', filtered_image)
    #Plot the sharpened image
    plt.subplot(1, 2, 2)
    plt.title("Sharpening")
    plt.imshow(sharpened_image)
    plt.show()
    
    
def noiseremoval(image, username):
    #Plot the original image
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    # Remove noise using a median filter
    filtered_image = cv2.medianBlur(image, 15)
    
    #Save the image
    # cv2.imwrite('Median-Blur_' + username +'.jpg', filtered_image)
    
    #Plot the blurred image
    plt.subplot(1, 2, 2)
    plt.title("Median Blur")
    plt.imshow(filtered_image)
    plt.show()
    
def imagescaling(image, username):
    #Plot the original image
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    # Scale the image by a factor of 2 along both axes
    scaled_image = cv2.resize(image, None, fx=2, fy=2)
    #Save the image
    #cv2.imwrite('scaled_' + username +'.jpg', scaled_image)
    #Plot the Scaled image
    plt.subplot(1, 2, 2)
    plt.title("Scaled")
    plt.imshow(scaled_image)
    plt.show()

def inversetransform(image, username):
    #Plot the original image
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    # Inversing the original image by subtracting from 255
    inverse_image = 255 - image
    
    #Save the image
    # cv2.imwrite('inverse_' + username +'.jpg', inverse_image)
    
    #Plot the Inverse image
    plt.subplot(1, 2, 2)
    plt.title("inverse Color")
    plt.imshow(inverse_image)
    plt.show()
    
    
#Processing function
def processing():
    username, matno = userinput(str(input("Username: ")), int(input("Matriculation Number: ")))
    images_list= os.listdir(directory)
    classmates = [im.split('.')[0] for im in images_list]
    if username in classmates:
        filename_ext = username + '.jpg'
        image = cv2.imread(filename_ext)
        print("\nWelcome " + username + "!\nBelow are the forms of image enhancement that you can perform.\nTo pick and enhancement please input the number that is next to the enhancement you want\n1. Visualise Code Spaces\n2. Addition of Image\n3. Adding Brightness and Contrast\n4. Sharpening\n5. Noise Removal\n6. Image Scaling\n7. Inverse Transformation.\n")
        number = int(input("Enter number for Enhancement: "))
        if number == 1:
            visualisecolorspaces(image, username)
        elif number == 2:
            additionofimage(image, username)
        elif number == 3:
            brightness_contrast(image, username)
        elif number == 4:
            sharpening(image, username)
        elif number == 5:
            noiseremoval(image, username)
        elif number == 6:
            imagescaling(image, username)
        elif number == 7:
            inversetransform(image, username)              
    else:
        print("\nUser not found.\nPlease try again.")
        
processing()
    
```
