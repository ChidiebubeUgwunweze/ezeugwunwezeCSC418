```python
import cv2
import os
from matplotlib import pyplot as plt
import numpy as np

directory = r'C:\Users\NEW\Documents\SCHOOL\Year_4_semester_2\CSC 418\Practical\week_3\Project\shyllon_images'
os.chdir(directory)
def waitandestroy():
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def sobel(image):
    # resize images
    image = cv2.resize(image, (500,400))

    #Apply Sabel edge detection
    sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 3)
    sobel_y = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 3)

    # Combine the gradient images
    sobel_combined = cv2.magnitude(sobel_x, sobel_y)

    # Display the original image and the Sobel edges
    cv2.imshow("Original", image)
    cv2.imshow("Sobel Edges", np.uint8(sobel_combined))
    waitandestroy()

def prewitt(image):
    image = cv2.resize(image, (500, 400))

    #Define Prewitt kernels
    kernel_x = np.array([[-1, -1, -1],
                         [0, 0, 0],
                         [1, 1, 1]])

    kernel_y = np.array([[-1, 0, 1],
                         [-1, 0, 1],
                         [-1, 0, 1]])

    # Apply Prewitt edge detection
    prewitt_x = cv2.filter2D(image, -1, kernel_x)
    prewitt_y = cv2.filter2D(image, -1, kernel_y)

    # Compute magnitude of gradients manually
    prewitt_combined = np.sqrt(np.square(prewitt_x) + np.square(prewitt_y))

    # Display the original image and the Prewitt edges
    cv2.imshow("Original", image)
    cv2.imshow("Prewitt Edges", np.uint8(prewitt_combined))
    waitandestroy()

def robert(image):
    image = cv2.GaussianBlur(image, (5, 5), 0)

    # Define Robert kernels
    kernel_x = np.array([[1, 0],
                         [0, -1]])

    kernel_y = np.array([[0, 1], 
                         [-1, 0]])

    # Apply Robert edge detection
    robert_x = cv2.filter2D(image, -1, kernel_x)
    robert_y = cv2.filter2D(image, -1, kernel_y)

    # Combine the gradient images
    robert_combined = np.sqrt(np.square(robert_x) + np.square(robert_y))

    # Display the original image and the Robert edges
    cv2.imshow("Original", image)
    cv2.imshow("Robert Edges", np.uint8(robert_combined))
    waitandestroy()

def laplacian(image):
    image = cv2.resize(image, (400,400))

    #Apply Laplacian edge detection
    laplacian = cv2.Laplacian(image, cv2.CV_64F)

    # Convert the output to an 8-bit image
    laplacian = cv2.convertScaleAbs(laplacian)

    #Display the original image and the laplacian edges
    cv2.imshow("Original", image)
    cv2.imshow("Laplacian Edges", laplacian)
    waitandestroy()

def canny(image):
    # resize images
    image = cv2.resize(image, (400, 400))

    # Perform Canny edge detection
    edges = cv2.Canny(image, 100, 200) # Adjust the threshold values as needed

    # Display the original image and the detected edges
    cv2.imshow("Original", image)
    cv2.imshow("Edges", edges)
    waitandestroy()

#Functions that checks if entry is valid or not
def usernamechecker():
    Usernames = ["Faustina", "George", "Desmond", "Charles"]
    Passwords = ["flashy67", "Talking12", "Parav68", "Tik345"]
    username = str(input("Username: "))
    password = str(input("Password: "))
    print("\n")
    for user, pass_ in zip(Usernames, Passwords):
        if username == user and password == pass_:
            process(username)
            return
    print("Entries not found please try again")
 
    
def process(username):
    #Welcome message
    print(f"Welcome {username} to the Yemisi Shyllon Meuseum Edge detection application.\nThis Meuseum's applcation has three categories you can pick images from.\nThey are listed below:\n\nI. Contemporary\nII. Modern\nIII. Traditional\n\n")
    category = input("Enter Category: ").lower()
    categories = os.listdir(directory)
    if category in categories:
        print(f"Below are the images in the {category} category:\n")
        images_list = os.listdir(f'{category}')
        image_noformat = [im.split('.')[0] for im in images_list]
        print("\n".join([f'{i+1}. {image_noformat[i]}' for i in range(len(image_noformat))]))
        image_name = input("Enter image of choice: ")
        print("\n")
        if image_name in image_noformat:
            photo = cv2.imread(f'{category}/{images_list[image_noformat.index(image_name)]}')
            print("Below are the Edge detections that you can perform:\n1. Sobel \n2. Prewitt \n3. Robert \n4. Laplacian \n5. Canny")
            detection = input("Enter type of edge detection: ")
            
            try:
                detection = int(detection)
                if detection == 1:
                    sobel(photo)
                elif detection == 2:
                    prewitt(photo)
                elif detection == 3:
                    robert(photo)
                elif detection == 4:
                    laplacian(photo)
                elif detection == 5:
                    canny(photo)
                else:
                    print("Please put in the right number or name")
            except ValueError:
                detection = detection.lower()
                if detection == "sobel":
                    sobel(photo)
                elif detection == "prewitt":
                    prewitt(photo)
                elif detection == "robert":
                    robert(photo)
                elif detection == "laplacian":
                    laplacian(photo)
                elif detection == "canny":
                    canny(photo)
                else:
                    print("Please put in the right number or name")
        else:
            print("Image not found please enter only the names you see.")
    else:
        print("Category not found please try again")
        
#Starting the application
usernamechecker()
```
